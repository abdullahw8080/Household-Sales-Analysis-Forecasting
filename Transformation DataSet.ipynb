```python
# sheet_name = "customers"
# sheet_name = "products"
# sheet_name = "employees"
sheet_name = "sales"

excel_file = pd.ExcelFile('Dataset_of_sales.xlsx')

df = pd.read_excel(excel_file, sheet_name=sheet_name)
```

```python

# إعداد الاتصال بقاعدة البيانات
# 'mysql+pymysql://user:password@host/database_name'
engine = create_engine('mysql+pymysql://root:@localhost/anlaysis_sales_datasets')

# رفع البيانات إلى جدول يحمل نفس اسم الورقة
# if_exists='replace' سيقوم بإنشاء الجدول أو استبداله إذا كان موجوداً
df.to_sql(sheet_name, con=engine, if_exists='replace', index=False)
print(f"تم رفع ورقة {sheet_name} بنجاح.")
```
